```python
import pyttsx3
import os
import sys
import wikipedia
import datetime
import webbrowser
import pyjokes
import speech_recognition as sr
import tkinter as tk
import threading
import win32gui, win32con

hide = win32gui.GetForegroundWindow()
win32gui.ShowWindow(hide , win32con.SW_HIDE)


engine=pyttsx3.init()
engine.setProperty('rate', 130)
engine.setProperty('volume',1.0)
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)


class speech_assistant(threading.Thread):
    def __init__(self):
        threading.Thread.__init__(self)
        self.start()
    
    def callback(self):
        self.root.quit()
    def run(self):
        global label
        self.root = tk.Tk()
        self.root.geometry("400x400")
        photo=tk.PhotoImage(file=r"Webp.net-resizeimage.png")
        f=tk.Label(image=photo)
        f.pack()
        self.root.protocol("WM_DELETE_WINDOW", self.callback)
        label=tk.Label(text="")
        label.pack()
        self.root.mainloop()
    
    def recognize(self):
        r=sr.Recognizer()
        with sr.Microphone() as source:
            label.config(text="Listening...")
            audio=r.listen(source)
            try:
                label.config(text="Recognizing...")
                query=r.recognize_google(audio,language='en-IN')
            except Exception as a:
                label.config(text=a)
                return "None"
            return query.lower()
        
    def execute(self):
            sub_schedule={14:'mobile computing and application devlopment',15:'mobile computing and application devlopment',16:'mobile computing and application devlopment',17:'mobile computing and application devlopment'\
                    ,18:'network management and administration',19:'network management and administration',20:'network management and administration'\
                        ,21:'advance java',22:'advance java',23:'advance java',24:'mobile computing and application devlopment',25:'mobile computing and application devlopment'\
                            ,26:'network management and administration',27:'network management and administration',28:'network management and administration'\
                                ,29:'advance java',30:'advance java'}
            
            command=self.recognize()
            if 'saman' in command:
                os.startfile(r'C:\Users\manav\Downloads\Best Sounds Your Dog Loves _ Sounds that attract dogs-104.5-132.1.mp3')
                
            if 'who are you' in command:
                engine.say('I am a voice assistant created by manav shah')
                engine.runAndWait()
            
            if 'joke' in command:
                joke=pyjokes.get_joke(language="en", category="neutral")
                engine.setProperty('rate', 100)
                engine.say(joke)
                engine.runAndWait()
                
            if 'you do' in command:
                engine.say('i am a doctor')
                engine.runAndWait()
                
            if 'greet me' in command:
                engine.say('Heyy there nice to meet you')
                engine.runAndWait()
            
            if "open youtube" in command:
                engine.say("Opening youtube")
                os.startfile(r"C:\Users\manav\OneDrive\Desktop\YouTube.lnk")
                engine.runAndWait()
            
            if 'subject' and 'today' in command:
                today=datetime.datetime.now().day
                engine.say(f"you have to prepare {sub_schedule[today]} today")
                engine.runAndWait()

            if 'tomorrow' in command:
                tommorrow=(datetime.datetime.now().day)+1
                engine.say(f"you have to prepare {sub_schedule[tommorrow]} tommorrow")
                engine.runAndWait()
                
            if 'schedule'.lower() in command:
                engine.say('you have paper of advance java on 31st may \
                    then you have network management and administration on 2nd june \
                        then mobile computing and application devlopment on 4th june ')
                engine.runAndWait()
                
            if "stop listening" in command:
                engine.say("Speech recognition is stopping now")
                engine.runAndWait()
                os.system("taskkill /im speech_assitant.exe")
                self.root.destroy()
                sys.exit("Speech Recognition is stopped")
            
            if "in wikipedia" in command:
                global label
                command=command.replace('wikipedia',"")
                results=wikipedia.summary(command,sentences=2)
                label.config(text=results)
            
            if "search" in command:
                command=command.replace('search','')
                #fire=webbrowser.get('google-chrome')
                #fire.open("What is one plus three")
                webbrowser.open(command)
            
            if "time" in command:
                time=datetime.datetime.now().strftime('%H:%M:%S')
                engine.say(time)
                engine.runAndWait()
                
            if 'open code for java' in command:
                os.system('cd D:/sem-6/Ajava/ajava_practicals') 
                os.system('code ..')
                os.system('exit')
                
            if 'open code for python' in command:
                os.system('D:\python_exercises\speech_assitant\python_programs') 
                os.system('code ..')
                os.system('exit')
            
            if 'open notes' in command:
                os.system('start D:\python_exercises\speech_assitant\python_programs\dist\exam_notes.exe')              
                
        
a=speech_assistant()
r=sr.Recognizer()
with sr.Microphone() as source:
    label.config(text="Please wait. Calibrating microphone...")
    # listen for 5 seconds and calculate the ambient noise energy level   
    r.adjust_for_ambient_noise(source,2)
    engine.say("Tell me what can i do")
    engine.runAndWait()
    while True:
        a.execute()
        
```
